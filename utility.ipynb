```
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/eggplant/data/data_with_img/deepface_train.csv')
df
```

```
df['tags'] = df['tags'].fillna('nothing')
df ['text'] = 'the image contains ' + df['tags'] + ' and he data is ' + df['text']
df = df.drop('tags',axis = 1)
df
```

```
df['tags'] = df['tags'].fillna('nothing')
for i in range (len(df)):
  if 'person' in df['tags'][i]:
    # print(df.iloc[i,:])
    df['tags'][i] = df['tags'][i].replace('person','')
    temp = df['tags'][i].split()
    df['tags'][i] = ' '.join(temp)
    df['text'][i] = 'the image contains ' + df['tags'][i] + ' a '+ str(df['race'][i]) + ' ' + str(df['emotion'][i]) + ' ' + str(df['gender'][i]) +  ' and the data is ' + df['text'][i]
  else:
    df ['text'][i] = 'the image contains ' + df['tags'][i] + ' and the data is ' + df['text'][i]
df
```

```
df = df.drop(['tags','age','gender','emotion','race'],axis = 1)
df
```
